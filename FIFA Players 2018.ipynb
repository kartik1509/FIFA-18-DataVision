```python
#Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
#Importing the database
raw_df = pd.read_csv(r'C:\Users\anand\Desktop\FIFA Players\fifa_players.csv')
raw_df
```

```python
filtered_df = raw_df.sort_values(by = 'overall_rating', ascending = False)
filtered_df.reset_index(inplace = True, drop = True)
filtered_df
```

```python
#Since we have a large number of stats we can group them to have a better understanding
filtered_df['Pace'] = (filtered_df['acceleration'] + filtered_df['sprint_speed'])/2
filtered_df['Shooting'] = (filtered_df['finishing'] + filtered_df['shot_power'] + filtered_df['long_shots'] + filtered_df['volleys'] + filtered_df['penalties'] + filtered_df['positioning'])/6
filtered_df['Passing'] = (filtered_df['vision'] + filtered_df['crossing'] + filtered_df['freekick_accuracy'] + filtered_df['short_passing'] + filtered_df['long_passing'] + filtered_df['curve'])/6
filtered_df['Dribbling'] = (filtered_df['agility'] + filtered_df['balance'] + filtered_df['reactions'] + filtered_df['ball_control'] + filtered_df['dribbling'] + filtered_df['composure'])/6
filtered_df['Defending'] = (filtered_df['interceptions'] + filtered_df['heading_accuracy'] + filtered_df['marking'] + filtered_df['standing_tackle'] + filtered_df['sliding_tackle'])/5
filtered_df['Physical'] = (filtered_df['jumping'] + filtered_df['stamina'] + filtered_df['strength'] + filtered_df['aggression'])/4
filtered_df
```

```python
#Rounding the newly created values to improve their significance
filtered_df['Pace'] = filtered_df['Pace'].astype(int)
filtered_df['Shooting'] = filtered_df['Shooting'].astype(int)
filtered_df['Passing'] = filtered_df['Passing'].astype(int)
filtered_df['Dribbling'] = filtered_df['Dribbling'].astype(int)
filtered_df['Defending'] = filtered_df['Defending'].astype(int)
filtered_df['Physical'] = filtered_df['Physical'].astype(int)
filtered_df
```

```python
#Lets see who the fastest and the slowest players are
pace_df = filtered_df.sort_values(by = 'Pace', ascending = False)
print('Fastest Players are:')
print(pace_df[['name','Pace']].head(2))
print('Slowest Players are:')
print(pace_df[['name','Pace']].tail(2))
```

```python
#Lets see which players have the best and worst shooting (while excluding the gks)
shooting_df = filtered_df.sort_values(by = 'Shooting', ascending = False)
shooting_df.drop(shooting_df[shooting_df['positions'] == 'GK'].index, inplace = True)
print('Player with best Shot:')
print(shooting_df[['name','Shooting']].head(1))
print('Player with worst Shot:')
print(shooting_df[['name','Shooting']].tail(1))
```

```python
#Lets identify the players with best and worst passing ability
passing_df = filtered_df.sort_values(by = 'Passing', ascending = False)
print('Player with best passing ability:')
print(passing_df[['name','Passing']].head(1))
print('Player with worst passing ability:')
print(passing_df[['name','Passing']].tail(1))
```

```python
#Lets identify the players with best and worst dribbling (while excluding the gks)
dribbling_df = filtered_df.sort_values(by = 'Dribbling', ascending = False)
dribbling_df.drop(dribbling_df[dribbling_df['positions'] == 'GK'].index, inplace = True)
print('Player with best dribbling:')
print(dribbling_df[['name','Dribbling']].head(1))
print('Player with worst dribbling:')
print(dribbling_df[['name','Dribbling']].tail(1))
```

```python
#Lets identify the players with best and worst defending (while excluding the gks)
defending_df = filtered_df.sort_values(by = 'Defending', ascending = False)
defending_df.drop(defending_df[defending_df['positions'] == 'GK'].index, inplace = True)
print('Player with best defending:')
print(defending_df[['name','Defending']].head(4))
print('Player with worst dribbling:')
print(defending_df[['name','Defending']].tail(1))
```

```python
#Lets find out who the strongest and the weakest players are
physical_df = filtered_df.sort_values(by = 'Physical', ascending = False)
print('Strongest Players:')
print(physical_df[['name','Physical']].head(3))
print('Weakest Players:')
print(physical_df[['name','Physical']].tail(2))
```

```python
ax = plt.subplot(polar = True)

plt.xticks(circle, attributes)

ax.set_rlabel_position(0)
plt.yticks([20,40,60,80,100], ['20','40','60','80','100'])
plt.ylim(0,100)

ax.plot(closed_circle, closed_values)
ax.fill(closed_circle, closed_values, alpha = 0.2)

plt.title('Leo Messi')
plt.show()
```
